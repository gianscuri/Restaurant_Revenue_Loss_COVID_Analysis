```python
vaccini = pd.read_csv(path_aggiuntivi + "somministrazioni-vaccini.csv", parse_dates=["data"])
# vaccini = pd.read_csv("Dati/somministrazioni-vaccini.csv", parse_dates=["data"])
vaccini
```

```python
### Eliminazione colonne non necessarie (elimino anche la divisione in maschi/femmine, non credo sia rilevante)

vaccini.drop(['area', 'm', 'f', 'N1', 'N2', 'ISTAT'], axis = 1, inplace=True)
```

```python
vaccini[vaccini['reg']=='Lombardia']
```

```python
### Mantengo solo Lombardia ed Emilia Romagna

vaccini_Lom_Emil = vaccini[vaccini['reg'].isin(['Lombardia', 'Emilia-Romagna'])]
vaccini_Lom_Emil
```

```python
### Sistemazione colonne

# Dose1: numero di prime dosi effettuate
# Dose2: numero di seconde dosi effettuate
# DoseUnica: numero di somministrazioni a soggetti con infezione pregressa, che completano il ciclo vaccinale con una sola dose
# Booster1: numero di somministrazioni di dose booster
# Booster3Dosi: numero di somministrazioni di dose booster a chi ha giÃ  il ciclo vaccinale primario con 3 dosi
# Booster2: numero di somministrazioni di seconda dose booster (120 giorni dopo la prima dose booster)

vaccini_Lom_Emil.columns = ['data', 'Fornitore', 'EtÃ ', 'Dose1', 'Dose2', 'DoseUnica', 'Booster1', 'Booster3Dosi', 
                            'Booster2', 'Regione']
vaccini_Lom_Emil
```

```python
### Elimino date dopo 2022-04-30

vaccini_Lom_Emil = vaccini_Lom_Emil[vaccini_Lom_Emil['data'] < '2022-05-01']
vaccini_Lom_Emil
```

```python
### Raggruppo per data e regione (non penso che serva la divisione in fasce d'etÃ  o fornitore per la questione ristorazione)

vaccini_Lom_Emil = vaccini_Lom_Emil.groupby(['data', 'Regione'], as_index=False).sum()
vaccini_Lom_Emil
```

```python
### Separo Lombardia ed Emilia

vacciniLombardia = vaccini_Lom_Emil[vaccini_Lom_Emil['Regione']=='Lombardia']
vacciniEmilia = vaccini_Lom_Emil[vaccini_Lom_Emil['Regione']=='Emilia-Romagna']
```

```python
### Creo colonne cumulate per tutte le colonne presenti, sia per Emilia Romagna che Lombardia

colonne = ['Dose1', 'Dose2', 'DoseUnica', 'Booster1', 'Booster3Dosi', 'Booster2']

for x in colonne:
    vacciniLombardia[str(x)+'Cum'] = vacciniLombardia[x].cumsum()
    vacciniEmilia[str(x)+'Cum'] = vacciniEmilia[x].cumsum()
```

```python
vacciniEmilia.head()
```

```python
vacciniLombardia.tail()
```

```python
### Converto 'Emilia-Romagna' in 'Emilia Romagna'

vacciniEmilia.Regione = vacciniEmilia.Regione.str.replace('-', ' ')
vacciniEmilia
```

```python
vacciniLombardia.to_csv("Dati\VacciniLombardia.csv")
vacciniEmilia.to_csv("Dati\VacciniEmiliaRomagna.csv")
```

```python
vaccini_Lom_Emil = pd.concat([vacciniEmilia, vacciniLombardia])
vaccini_Lom_Emil.sort_values('data', inplace=True)
vaccini_Lom_Emil.head(20)
```

```python
vaccini_rf = pd.read_csv("Dati/Dati per RANDOM FOREST/somministrazioni-vaccini-latest.csv", parse_dates=["data"])
vaccini_rf
```

```python
# Questi dati mi interessano solo per il ristorante 6 (R005), situato a Voghera, Lombardia
vaccini_rf.drop(['area', 'm', 'f', 'N1', 'N2', 'ISTAT'], axis = 1, inplace=True)
vaccini_rf = vaccini_rf[vaccini_rf['reg'].isin(['Lombardia'])]
vaccini_rf.columns = ['data', 'Fornitore', 'EtÃ ', 'Dose1', 'Dose2', 'DoseUnica', 'Booster1', 'Booster3Dosi', 
                            'Booster2', 'Regione']
```

```python
vaccini_rf = vaccini_rf.groupby(['data', 'Regione'], as_index=False).sum()
vaccini_rf
```

```python
### Creo colonne cumulate

colonne_rf = ['Dose1', 'Dose2', 'DoseUnica', 'Booster1', 'Booster3Dosi', 'Booster2']

for x in colonne_rf:
    vaccini_rf[str(x)+'Cum'] = vaccini_rf[x].cumsum()
```
